```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt

# Define the neural network architecture
class PINN(nn.Module):
    def __init__(self):
        super(PINN, self).__init__()
        # Neural network layers with increased neurons and hidden layers
        self.hidden_layer1 = nn.Linear(1, 50)
        self.hidden_layer2 = nn.Linear(50, 50)
        self.hidden_layer3 = nn.Linear(50, 50)
        self.hidden_layer4 = nn.Linear(50, 50)
        self.hidden_layer5 = nn.Linear(50, 50)
        self.hidden_layer6 = nn.Linear(50, 50)
        self.output_layer = nn.Linear(50, 1)
        self.activation = nn.Tanh()  # Activation function

    def forward(self, x):
        # Forward pass
        x = self.activation(self.hidden_layer1(x))
        x = self.activation(self.hidden_layer2(x))
        x = self.activation(self.hidden_layer3(x))
        x = self.activation(self.hidden_layer4(x))
        x = self.activation(self.hidden_layer5(x))
        x = self.activation(self.hidden_layer6(x))
        x = self.output_layer(x)
        return x

# Physics-Informed Loss Function
def pinn_loss(model, x_interior, x_boundary, u_boundary):
    # Predict temperature for interior points
    u_pred = model(x_interior)

    # Compute the second derivative using autograd for the physics-based loss
    u_x = torch.autograd.grad(u_pred, x_interior, grad_outputs=torch.ones_like(u_pred), create_graph=True)[0]
    u_xx = torch.autograd.grad(u_x, x_interior, grad_outputs=torch.ones_like(u_x), create_graph=True)[0]

    # Physics loss: Enforce the heat conduction equation (d^2u/dx^2 = 0)
    physics_loss = torch.mean(u_xx ** 2)

    # Boundary loss: Enforce boundary conditions
    u_boundary_pred = model(x_boundary)
    boundary_loss = torch.mean((u_boundary_pred - u_boundary) ** 2)

    # Total loss
    total_loss = physics_loss + boundary_loss
    return total_loss

# Main function to train the PINN
def train_pinn():
    # Generate training data
    L = 1.0  # Length of the rod
    x_interior = torch.linspace(0, L, 100).view(-1, 1).requires_grad_(True)  # Interior points (training data)
    x_boundary = torch.tensor([[0.0], [L]], requires_grad=True)  # Boundary points
    u_boundary = torch.tensor([[100.0], [0.0]])  # Boundary conditions (u(0) = 100, u(L) = 0)

    # Initialize the PINN model
    model = PINN()
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001)  # Optimizer

    # Training loop
    epochs = 100000
    for epoch in range(epochs):
        optimizer.zero_grad()  # Zero gradients

        # Compute loss
        loss = pinn_loss(model, x_interior, x_boundary, u_boundary)
        loss.backward()  # Backpropagation
        optimizer.step()  # Update weights

        # Print training progress
        if epoch % 1000 == 0:
            print(f"Epoch {epoch}, Loss: {loss.item()}")

    # Predict and plot results after training
    x_test = torch.linspace(0, L, 100).view(-1, 1)
    with torch.no_grad():
        u_pred = model(x_test)

    return x_test, u_pred

# Run the training
x_test, u_pred = train_pinn()

# Plot the results
plt.plot(x_test.detach().numpy(), u_pred.detach().numpy(), label='PINN Prediction')
plt.xlabel('Position along the rod (x)')
plt.ylabel('Temperature (u)')
plt.title('Temperature Distribution in a Rod using PINN')
plt.legend()
plt.show()
```
