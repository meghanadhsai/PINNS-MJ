```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt

# Define the neural network architecture with physics-informed features
class PINN(nn.Module):
    def __init__(self):
        super(PINN, self).__init__()
        # Neural network layers
        self.hidden_layers = nn.ModuleList([nn.Linear(1, 32)] + [nn.Linear(32, 32) for _ in range(5)])
        self.output_layer = nn.Linear(32, 1)
        self.activation = nn.ELU()  # Using ELU activation function

    def forward(self, x):
        # Neural network forward pass
        for layer in self.hidden_layers:
            x = self.activation(layer(x))
        return self.output_layer(x)

# Physics-Informed Loss Function with Adaptive Normalization
def pinn_loss(model, x_interior, x_boundary, u_boundary, normalization_factor, threshold=0.01):
    # Predict temperature for interior points
    u_pred = model(x_interior)

    # Compute the first and second derivatives using autograd for the physics-based loss
    u_x = torch.autograd.grad(u_pred, x_interior, grad_outputs=torch.ones_like(u_pred), create_graph=True)[0]
    u_xx = torch.autograd.grad(u_x, x_interior, grad_outputs=torch.ones_like(u_x), create_graph=True)[0]

    # Physics loss: Enforce the heat conduction equation (d^2u/dx^2 = 0)
    physics_loss = torch.mean(u_xx ** 2)

    # Boundary loss: Enforce boundary conditions
    u_boundary_pred = model(x_boundary)
    boundary_loss = torch.mean((u_boundary_pred - u_boundary) ** 2)

    # Adaptive normalization of loss terms
    ratio = physics_loss / (boundary_loss + 1e-8)  # Avoid division by zero
    if ratio > threshold:
        normalization_factor = 1.0
    else:
        normalization_factor = ratio / threshold

    # Total loss with adaptive normalization
    total_loss = normalization_factor * physics_loss + boundary_loss
    return total_loss, normalization_factor

# Main function to train the PINN
def train_pinn():
    # Generate training data
    L = 1.0  # Length of the rod
    x_interior = torch.linspace(0, L, 100).view(-1, 1).requires_grad_(True)  # Interior points (training data)
    x_boundary = torch.tensor([[0.0], [L]], requires_grad=True)  # Boundary points
    u_boundary = torch.tensor([[100.0], [0.0]])  # Boundary conditions (u(0) = 100, u(L) = 0)

    # Initialize the PINN model
    model = PINN()
    optimizer = torch.optim.AdamW(model.parameters(), lr=0.001)  # Optimizer (AdamW with weight decay)

    # Learning rate decay scheduler
    scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=2000, gamma=0.9)

    # Adaptive normalization factor initialization
    normalization_factor = 1.0

    # Training loop
    epochs = 50000
    for epoch in range(epochs):
        optimizer.zero_grad()  # Zero gradients

        # Compute loss
        loss, normalization_factor = pinn_loss(model, x_interior, x_boundary, u_boundary, normalization_factor)
        loss.backward()  # Backpropagation
        optimizer.step()  # Update weights
        scheduler.step()  # Adjust learning rate

        # Print training progress
        if epoch % 1000 == 0:
            current_lr = optimizer.param_groups[0]['lr']  # Get current learning rate
            print(f"Epoch {epoch}, Loss: {loss.item():.6f}, Normalization Factor: {normalization_factor:.6f}, LR: {current_lr:.6f}")

    # Predict and plot results after training
    x_test = torch.linspace(0, L, 100).view(-1, 1)
    with torch.no_grad():
        u_pred = model(x_test)

    return x_test, u_pred

# Run the training
x_test, u_pred = train_pinn()

# Plot the results
plt.plot(x_test.detach().numpy(), u_pred.detach().numpy(), label='PINN Prediction')
plt.xlabel('Position along the rod (x)')
plt.ylabel('Temperature (u)')
plt.title('Temperature Distribution in a Rod using PINN')
plt.legend()
plt.show()
```
